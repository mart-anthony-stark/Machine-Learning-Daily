# 2018-19 NBA Player Stats: Per Game EDA



## Web Scraping Data
 retrieve the "2018-19 NBA Player Stats: Per Game" data from http://www.basketball-reference.com/.

```python
import pandas as pd
import missingno

# Retrieve HTML table data
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
html = pd.read_html(url, header=0)
df2019  = html[0]
# Data Cleaning
raw = df2019.drop(df2019[df2019.Age == 'Age'].index)
raw
```

## Data Cleaning

### Data Dimension

```python
raw.shape
```

### Dataframe Contents

```python
raw.head()
```

### Checking Missing Values

```python
raw.isnull().sum()
```

```python
missingno.matrix(raw)
```

### Replace missing values with 0

```python
df = raw.fillna(0)
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Rk'], axis=1)
df
```

## Write to CSV

```python
df.to_csv('nba2k19.csv', index=False)
```
